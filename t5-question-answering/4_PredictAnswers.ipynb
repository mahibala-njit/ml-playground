```python
from  transformers  import  AutoTokenizer, AutoModelWithLMHead, pipeline

model_name = "MaRiOrOsSi/t5-base-finetuned-question-answering"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelWithLMHead.from_pretrained(model_name)
question = "What is 42?"
context = "42 is the answer to life, the universe and everything"
input = f"question: {question} context: {context}"
encoded_input = tokenizer([input],
                             return_tensors='pt',
                             max_length=512,
                             truncation=True)
output = model.generate(input_ids = encoded_input.input_ids,
                            attention_mask = encoded_input.attention_mask)
output = tokenizer.decode(output[0], skip_special_tokens=True)
print(output)
```

```python
def predict_answers(context, question):
  input = f"question: {question} context: {context}"
  encoded_input = tokenizer([input],
                             return_tensors='pt',
                             max_length=512,
                             truncation=True)
  output = model.generate(input_ids = encoded_input.input_ids,
                            attention_mask = encoded_input.attention_mask)
  output = tokenizer.decode(output[0], skip_special_tokens=True)
  return output
```

```python
def predict_answers_for_source(source):
  predicted_answer = []
  for index, row in source.iterrows():
    context = row[0] 
    original_keyword = row[1]
    find_begin = "<pad> question:"
    find_end = "</s>"
    question = row[2].replace(find_begin,'').replace(find_end,'')
    #print(context,original_keyword,question)
    answer = predict_answers(context,question)
    new_tuple = (context,original_keyword,question, answer)
    predicted_answer.append(new_tuple)
  return predicted_answer
```

```python
import pandas as pd
source=pd.read_excel("/content/gdrive/My Drive/CS 677 Project/dataset/context_questions.xlsx", sheet_name='Sheet_name_1')
```

```python
predicted_answer_tuple = predict_answers_for_source(source)
predicted_answer = pd.DataFrame(predicted_answer_tuple)
predicted_answer.to_excel("/content/gdrive/My Drive/CS 677 Project/dataset/predicted_answer.xlsx", sheet_name='Sheet_name_1')  
```
